```python
import pandas as pd
```

```python
fname = "/project/rivanna-stats/rivanna-corehours-2024-07-01T00:00:00-2024-07-31T23:59:59.csv"
df = pd.read_csv(fname, delimiter="|")
df = df[df["start"]!="Unknown"]
```

```python
df["start"] = pd.to_datetime(df["start"], utc=False) #-pd.to_datemine(df["submit"], utc=True)
df["submit"] = pd.to_datetime(df["submit"], utc=False)
df["end"] = pd.to_datetime(df["end"], utc=False)
```

```python
df.dtypes
```

```python
df["wait"] = df["start"]-df["submit"] 
df["runtime"] = df["end"]-df["start"]
```
